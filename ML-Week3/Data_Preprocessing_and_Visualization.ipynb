**Subplot**

With the use of subplot method, we can create many subplots in a figure. The subplot() method takes three arguments viz are nrows, ncols, and index.

```python
plt.subplot(1,2,1) # 1 row, 2 columns, index 1
plt.plot([1,2,3,4,5], [2,4,6,8,10], 'g')
plt.title('First')

plt.subplot(1,2,2) # 1 row, 2 columns, index 2
plt.plot([1,3,6,9,12], [3,6,9,12,15], 'b')
plt.title('Second')

plt.subplots_adjust(hspace=0.4) # for adjusting spaces inbetween subplots

plt.suptitle('Two subplots')
plt.show()
```

##### Dealing with missing values

Impute Missing Values

Imputing refers to using a model to replace missing values.

There are many options we could consider when replacing a missing value, for example:

A constant value that has meaning within the domain, such as 0, distinct from all other values.
A value from another randomly selected record.
A mean, median or mode value for the column.
A value estimated by another predictive model.

or you can delete the row or column base on the nature of the dataset

```python
titanic_df = pd.read_csv('titanic_train.csv')
titanic_df.head()
```

```python
def calc_missing(data):
    percent_missing = round(data.isna().sum() * 100 / len(data), 2)
    missing_count = data.isna().sum()
    dtype_list = data.dtypes
    column_name_list = list(data.columns)
    missing_df = pd.DataFrame(zip(missing_count, percent_missing,dtype_list, column_name_list),
                              columns=['Missing_Counts', '%Missing', 'DataType', 'ColumnName'])
    missing_df.set_index('ColumnName', inplace=True)
    return missing_df
```

```python
calc_missing(titanic_df)
```

```python
def missing(data):
    count_missing = data.isnull().sum()
    count_missing_percentage = round((data.isnull().sum() * 100) / len(data),)
    missing_data_type = data.dtypes
    missing_column_name = data.columns
    missing_df = pd.DataFrame(zip(count_missing, count_missing_percentage, missing_data_type,
                                 missing_column_name), columns=['Missing Count', '%Missing',
                                                               'Missing Dtype', 'ColumnName'])
    missing_df = missing_df.set_index('ColumnName')
    return missing_df
```

```python
missing(titanic_df)
```
